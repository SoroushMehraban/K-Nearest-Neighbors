```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from knn.knn import KNearestNeighbor
import knn.dataloader as dataloader
import time
```

# Iris dataset

```python
dataset = dataloader.get_iris_dataset()
dataset.shape
```

```python
k_values = [2, 3, 5, 10, 12, 15, 20, 50]
accuracy_values = []
for k in k_values:
    knn = KNearestNeighbor(dataset, k)
    accuracy_values.append(knn.predict())
```

```python
fig, ax = plt.subplots()
ax.plot(k_values, accuracy_values,
        marker='o')
ax.set(xlabel="K", ylabel='Accuracy');
```
